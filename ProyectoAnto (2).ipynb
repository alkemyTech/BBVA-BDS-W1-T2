# **BBT2-27** - Filtrar datos, reducir volumen de orders_items y sellers

```python
query_01 = '''
WITH ItemRevenue AS (
    SELECT
        oi.*,
        s.seller_id,
        oi.price * oi.freight_value AS total_revenue,
        ROW_NUMBER() OVER(PARTITION BY s.seller_id ORDER BY oi.price * oi.freight_value DESC) AS row_num
    FROM
        order_items oi
    INNER JOIN
        sellers s ON oi.seller_id = s.seller_id
)

SELECT
    *
FROM
    ItemRevenue
WHERE
    row_num <= 3;
'''
top_items_by_seller = pd.read_sql_query(query_01, conn)
# Eliminar registros duplicados (si existen)
top_items_by_seller = top_items_by_seller.drop_duplicates()
top_items_by_seller.head()
```

```python
query_28 = '''
WITH ItemRevenue AS (
    SELECT
        oi.*,
        p.product_category_name,
        oi.price * oi.freight_value AS total_revenue,
        ROW_NUMBER() OVER(PARTITION BY p.product_category_name ORDER BY oi.price * oi.freight_value DESC) AS row_num
    FROM
        order_items oi
    INNER JOIN
        products p ON oi.product_id = p.product_id
    INNER JOIN
        orders o ON oi.order_id = o.order_id
)

-- Select the top three items by total revenue per category
SELECT
    ItemRevenue.*,
    o.order_purchase_timestamp
FROM
    ItemRevenue
INNER JOIN
    orders o ON ItemRevenue.order_id = o.order_id
WHERE
    row_num <= 3;
'''
top1_items_by_category = pd.read_sql_query(query_28, conn)
top1_items_by_category.head()
```

## **BBT2-67** (ANTO)- Querys en spark SQL, análisis C

Generar una consulta que cumpla los siguientes puntos:
Totalizar por las órdenes más vendidas por ciudad del vendedor, año y mes limite de entrega sumando monto de envio y promedio de monto de envio.

PARA: Enriquecer el análisis de los datos según requerimiento. Para entender las frecuencia de compra y la incidencia de los envios por ciudad del vendedor.

```python
# Consulta para totalizar por las órdenes más vendidas por ciudad del vendedor, año y mes de entrega,
# y sumar el monto de envío y el promedio de monto de envío.
query_C = """
SELECT
    seller_id AS Vendedor,
    YEAR(shipping_limit_date) AS Anio_Limite_Entrega,
    MONTH(shipping_limit_date) AS Mes_Limite_Entrega,
    COUNT(DISTINCT order_id) AS Total_Ordenes,
    SUM(price + freight_value) AS Total_Monto_Envio,
    AVG(price + freight_value) AS Promedio_Monto_Envio
FROM
    vt_Analysis_C
GROUP BY
    seller_id, Anio_Limite_Entrega, Mes_Limite_Entrega
ORDER BY
    Total_Ordenes DESC
LIMIT 10
"""

# Ejecuta la consulta y muestra los resultados
query_C = spark.sql(query_C)
query_C.show()
```

Gráfico de barras para los 10 vendedores principales por total de órdenes:

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Total_Monto_Envio', y='Vendedor', data=query_C.toPandas(), color='skyblue', label='Total Monto de Envío')
sns.barplot(x='Promedio_Monto_Envio', y='Vendedor', data=query_C.toPandas(), color='orange', label='Promedio Monto de Envío')
plt.xlabel('Monto de Envío')
plt.ylabel('Vendedor')
plt.title('Total y Promedio de Monto de Envío por Vendedor')
plt.legend()
plt.show()
```

Gráfico de barras apiladas para el total y el promedio de monto de envío por vendedor:

```python
query_C_pandas = query_C.toPandas()
query_C_pandas_sorted = query_C_pandas.sort_values(by='Total_Monto_Envio', ascending=False)

# Create a new column with vendor labels 'A', 'B', 'C', etc.
query_C_pandas_sorted['Vendor_Label'] = [chr(ord('A') + i) for i in range(len(query_C_pandas_sorted))]

plt.figure(figsize=(12, 6))
sns.barplot(x='Total_Monto_Envio', y='Vendor_Label', data=query_C_pandas_sorted, color='skyblue', label='Total Monto de Envío', palette='Blues_d')
plt.xlabel('Monto de Envío')
plt.ylabel('Vendedor')
plt.title('Total de Monto de Envío por Vendedor')
plt.legend()
plt.show()
```

Gráfico de líneas para el total de órdenes a lo largo del tiempo (meses):

```python
promedio_monto_envio = query_C.toPandas()

plt.figure(figsize=(12, 6))
plt.plot(promedio_monto_envio['Anio_Limite_Entrega'].astype(str) + '-' + promedio_monto_envio['Mes_Limite_Entrega'].astype(str), promedio_monto_envio['Promedio_Monto_Envio'], marker='o')
plt.xlabel('Año-Mes de Entrega')
plt.ylabel('Promedio de Monto de Envío')
plt.title('Promedio de Monto de Envío a lo largo del Tiempo')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

## **BBT2-68** (ANTO)- Querys en spark SQL, análisis D

Generar una consulta que cumpla los siguientes puntos:
Totalizar por las órdenes más vendidas por categoria, por año y mes limite de entrega sumando monto de envio y promedio de monto de envio.

PARA: Enriquecer el análisis de los datos según requerimiento. Para entender las frecuencia de compra y la incidencia de los envios por categoria de los productos.

```python
query_D = '''
SELECT
    product_category_name AS categoria_del_producto,
    YEAR(shipping_limit_date) AS ano_limite_entrega,
    MONTH(shipping_limit_date) AS mes_limite_entrega,
    COUNT(order_id) AS total_ordenes,
    SUM(freight_value) AS total_monto_envio,
    AVG(freight_value) AS promedio_monto_envio
FROM
    vt_Analysis_D
GROUP BY
    product_category_name,
    YEAR(shipping_limit_date),
    MONTH(shipping_limit_date)
ORDER BY
    total_ordenes DESC
'''
query_D = spark.sql(query_D)
query_D.show(10)
```

Gráfico de barras apiladas para el total y el promedio de monto de envío por categoría de producto:

```python
query_D_pandas = query_D.toPandas()

query_D_pandas_sorted = query_D_pandas.sort_values(by='total_monto_envio', ascending=False)

top_10_categorias = query_D_pandas_sorted.head(10)

plt.figure(figsize=(12, 6))

# Crear un gráfico de barras apiladas para el top 10
sns.barplot(x='total_monto_envio', y='categoria_del_producto', data=top_10_categorias, color='skyblue', label='Total Monto de Envío')
sns.barplot(x='promedio_monto_envio', y='categoria_del_producto', data=top_10_categorias, color='orange', label='Promedio Monto de Envío')
plt.xlabel('Monto de Envío')
plt.ylabel('Categoría de Producto')
plt.title('Top 10 Categorías de Producto por Total y Promedio de Monto de Envío')
plt.legend()

plt.show()
```

## **BBT2-73** (ANTO)- Querys en spark SQL, análisis I

Generar una consulta que cumpla los siguientes puntos:
Calcular el maximo monto mensual por categoria de producto y estado de envío, agrupado por año y mes de envio.

PARA: Enriquecer el análisis de los datos según requerimiento. Para entender el comportamiento de compra de los productos y su estado de envío mensual.

```python
# Consulta SQL para calcular el máximo monto mensual por categoría de producto y estado de envío
query_I = '''
SELECT
    YEAR(C.shipping_limit_date) AS Anio_Envio,
    MONTH(C.shipping_limit_date) AS Mes_Envio,
    D.product_category_name AS Categoria_Producto,
    I.translated_category AS Estado_Envio,
    MAX(C.price) AS Max_Monto_Mensual
FROM
    vt_Analysis_C C
JOIN
    vt_Analysis_D D
ON
    C.product_id = D.product_id
JOIN
    vt_Analysis_I I
ON
    C.order_id = I.order_id
GROUP BY
    Anio_Envio, Mes_Envio, Categoria_Producto, Estado_Envio
ORDER BY
    Anio_Envio, Mes_Envio, Categoria_Producto, Estado_Envio;

'''
query_I = spark.sql(query_I)
query_I.show()
```

Gráfico de líneas para seguir el máximo monto mensual por categoría de producto a lo largo del tiempo (meses):

```python
query_I_pandas = query_I.toPandas()

# Luego, ordena el DataFrame por la columna Max_Monto_Mensual de manera descendente
query_I_pandas_sorted = query_I_pandas.sort_values(by='Max_Monto_Mensual', ascending=False)

# Selecciona solo las 20 primeras filas (los 20 principales)
top_20_query_I_pandas = query_I_pandas_sorted.head(20)

# Crea el gráfico de líneas con los 20 principales resultados
plt.figure(figsize=(12, 6))
sns.lineplot(x='Mes_Envio', y='Max_Monto_Mensual', hue='Categoria_Producto', data=top_20_query_I_pandas)
plt.xlabel('Mes de Envío')
plt.ylabel('Máximo Monto Mensual')
plt.title('Top 20 Categorías de Producto por Máximo Monto Mensual a lo largo del Tiempo')
plt.legend(title='Categoría de Producto')
plt.show()
```
