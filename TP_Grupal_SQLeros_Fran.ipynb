##Crear un DataFrame que contenga el JOIN de la tabla customers y tabla orders. Extraer los datos de clientes con sus ordenes

```python
#Crear un DataFrame que contenga el JOIN de la tabla customers y tabla orders.
#Extraer los datos de clientes con sus ordenes

customers = pd.read_sql_query(
    '''
    SELECT c.customer_unique_id,o.order_id
    FROM customers c
    JOIN orders o ON
    o.customer_id=c.customer_id
    GROUP BY c.customer_id
    ''',
    conn
)
customers
```

## Crear un DataFrame que contenga el JOIN de la tabla orders y tabla order_items. Extraer los datos de ordenes con sus items

```python
#Crear un DataFrame que contenga el JOIN de la tabla orders y tabla order_items.
#Extraer los datos de ordenes con sus items

orders = pd.read_sql_query(
    '''
    SELECT oi.order_id, product_id
    FROM orders o
    JOIN order_items oi ON
    oi.order_id=o.order_id
    GROUP BY oi.order_id, product_id
    ''',
    conn
)
orders
```

## Crear un DataFrame que contenga el JOIN de la tabla orders, tabla order_payments y customers. Extraer los datos de clientes con sus m√©todos de pagos por ciudad.

```python
#Crear un DataFrame que contenga el JOIN de la tabla orders, tabla order_payments y customers.
#Extraer los datos de clientes con sus m√©todos de pagos por ciudad.

payment_type_client = pd.read_sql_query(
    '''
    SELECT c.customer_unique_id,c.customer_city,op.payment_type
    FROM customers c
    JOIN orders o ON
    c.customer_id=o.customer_id
    JOIN order_payments op ON
    op.order_id=o.order_id
    GROUP BY c.customer_id,op.payment_type
    ''',
    conn
)

#orders tiene order_id y customer_id
#customer tiene customer_id, customer_unique_id y customer_city y customer_state
#order_payments tiene order_id, payment_type

payment_type_client
```

```python
##Ejercicio de Meli

## Crear un DataFrame que contenga el JOIN de la tabla orders y tabla geolocation

products = pd.read_sql_query(
    '''
    SELECT c.customer_id,o.order_id,g.geolocation_lat,g.geolocation_lng
    FROM customers c
    JOIN orders o ON
    o.customer_id=c.customer_id
    JOIN geolocation g ON
    g.geolocation_zip_code_prefix=c.customer_zip_code_prefix
    GROUP BY c.customer_id
    LIMIT 10
    ''',
    conn
)

# orders tiene order_id y customer_id
# geolocation tiene geolocation_zip_code_prefix, geolocation_city	y geolocation_state
products
```
