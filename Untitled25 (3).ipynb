# **BBT2-17** - Extraccion de datos desde la tabla orders_items y sellers

# **BBT2-22** - Extracción de datos desde las tablas orders y reviews

```python
query_3 = """
SELECT o.*, r.review_score
FROM orders o
INNER JOIN reviews r ON o.order_id = r.order_id
"""
orders_reviews_df = pd.read_sql_query(query_3, conn)
orders_reviews_df
```

# **BBT2-23** - Extracción de datos desde las tablas orders_items, products y products_category_translation

```python
query_2 = '''
SELECT oi.*, p.product_category_name
FROM order_items oi
INNER JOIN products p ON oi.product_id = p.product_id
INNER JOIN product_category_name_translation t ON p.product_category_name = t.product_category_name
'''

product_category_name = pd.read_sql_query(query_2, conn)
product_category_name
```

# **BBT2-27** - Filtrar datos, reducir volumen de orders_items y sellers

```python
query_01 = '''
WITH ItemRevenue AS (
    SELECT
        oi.*,
        s.seller_id,
        oi.price * oi.freight_value AS total_revenue,
        ROW_NUMBER() OVER(PARTITION BY s.seller_id ORDER BY oi.price * oi.freight_value DESC) AS row_num
    FROM
        order_items oi
    INNER JOIN
        sellers s ON oi.seller_id = s.seller_id
)

SELECT
    *
FROM
    ItemRevenue
WHERE
    row_num <= 3;
'''
top_items_by_seller = pd.read_sql_query(query_01, conn)
top_items_by_seller
```

# **BBT2-28** -  Filtrar y reducir volumen de orders_items y products,  para resguardar los 3 ítems de mayor importe.

```python
query_02 = '''
WITH ItemRevenue AS (
    SELECT
        oi.*,
        p.product_category_name,
        oi.price * oi.freight_value AS total_revenue,
        ROW_NUMBER() OVER(PARTITION BY p.product_category_name ORDER BY oi.price * oi.freight_value DESC) AS row_num
    FROM
        order_items oi
    INNER JOIN
        products p ON oi.product_id = p.product_id
)

-- Seleccionar los tres ítems de mayor importe por categoría
SELECT
    *
FROM
    ItemRevenue
WHERE
    row_num <= 3;
'''
top_items_by_category = pd.read_sql_query(query_02, conn)
top_items_by_category
```
